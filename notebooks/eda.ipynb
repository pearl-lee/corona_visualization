## 데이터 살펴보기

```python
import re

selng = card.selng_cascnt.unique()
x = re.compile('[가-힣]+')

# x.match(selng[0])
for a in selng:
    m = x.match(a)
    if m:
        print(m.group())
```

```python
notint = ['석재', '커텐', '영상물', '복지매장', '축협직영매장']
card_0 = card[card.selng_cascnt == notint[0]]
card_1 = card[card.selng_cascnt == notint[1]]
card_2 = card[card.selng_cascnt == notint[2]]
card_3 = card[card.selng_cascnt == notint[3]]
card_4 = card[card.selng_cascnt == notint[4]]
```

```python
len(card_0) + len(card_1) + len(card_2) + len(card_3) + len(card_4)
```

```python
card1 = pd.concat([card_0, card_1, card_2, card_3, card_4])
```

```python
card2 = pd.concat([card, card1])
card2.drop_duplicates(keep=False, inplace=True)
```

```python
len(card1) + len(card2)
```

```python
card2 = card2.astype({'selng_cascnt': int, 'salamt': int})
```
